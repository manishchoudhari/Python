```python
# Special operator with list
# List concatenate with +
list1=[1,2,3]
list2=[4,5,6]
print(list1+list2)
```

```python
# Special operator with list
# List Repetition with *
list1=[1,2,3]
print(list1*3)
```

```python
# Special operator with list
# Membership: in / not in
print(5 in list1)
print(3 in list1)
```

```python
# Special operator with list
# Membership: in / not in
print(5 not in list1)
print(3 not in list1)
```

```python
# Special operator with list
# Slicing
# Slicing: list[start:end:step]
list = [1,5,3,0,7,8,11] # start at index 1, end(exclude) at index 5
print(list[1:5])
```

```python
# Special operator with list
# Slicing
# Slicing: list[start:end:step]
list = [1,5,3,0,7,8,11] # start at index 1, end(exclude) at index 5 and step every 2 element
print(list[1:5:2])
```
